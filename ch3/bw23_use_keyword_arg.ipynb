```python
def remainder(number, divisor):
    return number % divisor

assert remainder(20, 7) == 6
```

```python
remainder(20, 7)
remainder(20, divisor=7)
remainder(number=20, divisor=7)
remainder(divisor=7, number=20)
```

```python
remainder(number=20, 7)
```

```python
remainder(20, number=20)
```

```python
my_kwargs = {'number': 20, 'divisor': 7}

assert remainder(**my_kwargs) == 6
```

```python
my_kwargs = {'divisor': 7}

assert remainder(20, **my_kwargs) == 6
```

```python
my_kwargs = {'divisor': 7}

other_kwargs = {'number': 20}
assert remainder(**my_kwargs, **other_kwargs) == 6
```

```python
def print_parameters(**kwargs):
    for key, value in kwargs.items():
        print(f'{key} = {value}')
```

```python
print(print_parameters(alpha=1.5, beta=9, 감마=4))
```

```python
def flow_rate(weight_diff, time_diff, period=1):
    return (weight_diff / time_diff) * period
```

```python
weight_diff = 0.5
time_diff = 3
flow_rate(weight_diff, time_diff)
```

```python
flow_rate(weight_diff, time_diff, period=3600)
```

```python
def flow_rate(weight_diff, time_diff, period=1, units_per_kg=1):
    return (weight_diff * units_per_kg / time_diff) * period
```

```python
pounds_per_hour = flow_rate(weight_diff, time_diff, period=3600, units_per_kg=2.2)
print(pounds_per_hour)
```

```python
pounds_per_hour = flow_rate(weight_diff, time_diff, 3600, 2.2)
print(pounds_per_hour)
```
