```python
import logging
logging.getLogger().setLevel(logging.WARNING)
def my_function():
    logging.debug('디버깅 데이터')
    logging.error('이 부분을 오류 로그')
    logging.debug('추가 디버깅 데이터')
```

```python
my_function()
```

```python
from contextlib import contextmanager

@contextmanager
def debug_logging(level):
    logger = logging.getLogger()
    old_level = logger.getEffectiveLevel()
    logger.setLevel(level)

    try:
        yield
    finally:
        logger.setLevel(old_level)
```

```python
with debug_logging(logging.DEBUG):
    print('* 내부:')
    my_function()

print("* 외부:")
my_function()
```

```python
with open('my_output.txt', 'w') as handle:
    handle.write('데이터 입니다.!')
```

```python
@contextmanager
def log_level(level, name):
    logger = logging.getLogger(name)
    old_level = logger.getEffectiveLevel()
    logger.setLevel(level)
    try:
        yield logger
    finally:
        logger.setLevel(old_level)
```

```python
with log_level(logging.DEBUG, 'my-log') as logger:
    logger.debug(f'대상: {logger.name}!')
    logging.debug('이 메세지는 출력되지 않습니다.')
```

```python
logger = logging.getLogger('my-log')
logger.debug('디버그 메세지는 출력되지 않습니다.')
logger.error('오류 메세지는 출력됩니다.')
```
